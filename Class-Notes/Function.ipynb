#### Create a function that takes an integer as input and returns "Even" if it's even and "Odd" if it's odd.

```python
def even_odd_checker(n):
    if n%2==0:
        print("even")
    else:
        print("odd")
```

```python
even_odd_checker(45)
```

```python
#0,1,1,2,3,5,8......
fibo=[0,1]  #initialization[0,1,1,2,3]
for i in range(1,19):
    next_term=fibo[-1]+fibo[-2]
    fibo.append(next_term)
print(fibo)
```

```python
def fibbo(n):
    fibo=[0,1]  #initialization[0,1,1,2,3]
    for i in range(1,n+1):
        next_term=fibo[-1]+fibo[-2]
        fibo.append(next_term)
    print(fibo)

fibbo(5)
```

```python
even=[]
odd=[]
def filter_even_odd(l):
    for i in l:
        if i%2==0:
            if i not in even:
                even.append(i)
        else:
            if i not in odd:
                odd.append(i)
    return even,odd
filter_even_odd([2,4,5,7,9,2,5,7,9,0,7])            
```

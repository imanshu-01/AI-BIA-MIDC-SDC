# 5.Identifying Palindromes:

```python
words = ["level", "hello", "radar", "python","nayan","level"]
palindromes=[]
for w in words:
    if w == w[::-1]:   #forward word==reverse of that word
        palindromes.append(w)  
palindromes
```

```python
#using a for loop
# Define the string
s = "Hello python world, welcome to the world of programming"

# Input from user
find = str(input("Enter the word you want to find: "))  #Hello

# Split the string into words
words = s.split()
#print(words)
# Flag to check if word is found
found = False

# Iterate through each word in the list
for word in words:
    if word == find:
        print(find, "Found in the string")
        found = True
        break  # Exit the loop once the word is found

# If the word is not found, print a message
if not found:
    print(find, "Not found in the string")
```

# Prime number 

```python
num=int(input("Enter number:"))
if num==1:
    print("not a prime number")
else:
    for i in range(2,num):#(2,7)
        if num%i==0:
            print("not a prime number")
            break
       
    else:
        print("prime")

            
```

```python
num = int(input("Enter number:"))

if num == 1:
    print("Not a prime number")
else:
    for i in range(2, int(num ** 0.5) + 1): #(2,7) # Only check up to the square root of num
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime")
```

```python
# Get number of rows from user
rows = int(input("Enter the number of rows: "))

i = 1  # Outer loop counter (for each row)
while i <= rows:
    # Print spaces
    j = 1
    while j <= rows - i:
        print(" ", end="")
        j += 1
    
    # Print stars
    k = 1
    while k <= (2 * i - 1):
        print("*", end="")
        k += 1

    print()  # Move to the next line
    i += 1
```

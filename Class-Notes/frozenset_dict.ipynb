### Indexing and Slicing 

```python
d = {"name": "Aditya", "age": 25, "city": "Mumbai", "country": "India"}

# Convert the keys to a list and slice it
keys_list = list(d.keys())
keys_list
```

```python
print(keys_list[:4])  # Output: ['name', 'age']
```

```python
# Convert the values to a list and slice it
values_list = list(d.values())

values_list
```

```python
print(values_list[:3])  # Output: ['Aditya', 25]
```

```python
# Convert items to a list of key-value pairs (tuples)
items_list = list(d.items())
items_list
```

```python
print(items_list[:2])  # Output: [('name', 'Aditya'), ('age', 25)]
```

```python
d = {"name": "Aditya", "age": 25, "city": "Mumbai", "country": "India"}

# Slice first two items
items_list = list(d.items())[:2]
print(items_list)  # Output: [('name', 'Aditya'), ('age', 25)]
```

### Example code to calculate discount based on purchase amount and membership

```python
purchase_amount = float(input("Enter purchase amount: "))
is_member = str(input("Are you a member? (yes/no): ")).lower()
mo_no=9987845110
# mono=int(input("please enter ur mobile no.= "))
if is_member == "yes":
    mono=int(input("please enter ur mobile no.= "))
    if  mono==mo_no:
        if purchase_amount >= 100:
            discount = 0.1 * purchase_amount  #10%
        else:
            discount = 0.05 * purchase_amount  #5%
    
else:
    if purchase_amount >= 100:
        discount = 0.02 * purchase_amount  #2%
    else:
        discount = 0

paid_amount = purchase_amount - discount
print("Total amount after discount:", paid_amount)
```

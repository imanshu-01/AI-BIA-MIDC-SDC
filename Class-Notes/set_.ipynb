## Operators

- union():combine
- intersection(): common elements
- difference(): elements present in set1 but not in set2
- symmetric_difference():uncommon elements


```python
s1={1,2,4,6,8,1,34,6}
s2={"a","z",1,2,5,9,2}
```

```python
#union()/|
s1.union(s2)
```

```python
s1|s2 #union
```

```python
s1.intersection(s2)
```

```python
s1&s2
```

```python
s1={1,2,4,6,8,1,34,6}  # difference{4,6,8,34}
s2={"a","z",1,2,5,9,2}  # symmetric_difference{4,6,8,34, a,z,5,9}
```

```python
s1
```

```python
s2
```

```python
s1.difference(s2)
```

```python
s2.difference(s1)
```

```python
s1-s2
```

```python
s2-s1
```

```python
s1.symmetric_difference(s2)
```

```python
s1^s2
```

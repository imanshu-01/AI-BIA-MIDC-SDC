### Recurssive function
A recursive function is a function that calls itself in order to solve a problem. 

```python
n=5  #5*4*3*2*1   #without function
fact=1
for i in range(1,n+1):  #(1,6)==>i=1,2,3,4,5
    fact=fact*i
print(fact)
```

```python
def factorial(n):  #without recurssion but using function
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
```

```python
factorial(5)
```

```python
factorial(7)
```

```python
def fact(n):
    factorial=1
    while n>1: #5>1
        factorial=factorial*n
        n=n-1  #5-1==>4,4-1==>3
    return factorial   

#5*4*3*2*1
```

```python
fact(5)
```

### Factorial with recurssion

```python
#with recurssion
def factorial(n):
    if n==1 or n==0:
        return 1
    else:
        return (n*factorial(n-1))
    
#factorial(4)
#4*factorial(3)
#4*3*factorial(2)
#4*3*2*factorial(1)
#4*3*2*1
```

```python
factorial(4)
```

```python
factorial(5)
```

```python
#n*factorial(n-1)
4*factorial(3)
4*3*2*1
```

```python
factorial(5)
```

```python
#with function but without recurssion
def fib(n):
    fibo=[0,1]
    for i in range(1,n-1):
        next_term=fibo[-1]+fibo[-2]
        fibo.append(next_term)
    return fibo 
fib(5)
```

```python
##without recurssion
#[0,1,1,2,3,5,8,13.....]
def fibonacci(n):
    fib_sequence = [0, 1]  # Initialize the sequence with the first two terms

    if n <= 1:
        return fib_sequence

    for i in range(2, n ):  #(2,5)
        next_term = fib_sequence[i - 1] + fib_sequence[i - 2]
        fib_sequence.append(next_term)

    return fib_sequence

fibonacci(5)

#f[1]+[0]=1+0=1
#f[2]+f[1]=1+1=2
#f[3]+f[2]=2+1=3

```

```python
# with recurssion
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
fibonacci(4)
for i in range(0,4):  # Print the first 5 terms of the sequence
    print(fibonacci(i), end=", ")    
    
    
    
#0,1,
#f(1)+f(0)==1+0=1
#f(2)+f(1)
#f(1)+f(0)+f(1)+1+0+1=2
```

```python
 how the function computes fibonacci(4):

fibonacci(4)
Calls fibonacci(3) and fibonacci(2)

fibonacci(3)
Calls fibonacci(2) and fibonacci(1)

fibonacci(2)
Calls fibonacci(1) and fibonacci(0)

fibonacci(1) and fibonacci(0)
Directly return 1 and 0, respectively.

Back to fibonacci(2):
Returns 1 (from fibonacci(1)) + 0 (from fibonacci(0)) = 1

Back to fibonacci(3):
Calls fibonacci(2) again (already computed as 1) and fibonacci(1)

Returns 1 (from fibonacci(2)) + 1 (from fibonacci(1)) = 2

Back to fibonacci(4):
Calls fibonacci(3) (already computed as 2) and fibonacci(2) (already computed as 1)

Returns 2 (from fibonacci(3)) + 1 (from fibonacci(2)) = 3
```

### Lambda with filter

```python
l=[56,33,13,23,22,45,56,67,3,2,3,4]
x=list(filter(lambda a:a%2==0,l))  #filter elements from a sequence based on a condition.
print(x)
```

```python
def f1(p):
    a1=[]
    for i in p:
        if i<6:
            a1.append(i)
    return a1
f1([56,33,13,23,22,45,56,67,3,1,2,5,6])
```

```python
p=[56,33,13,23,22,45,56,67,3,1,2,5,6]
x=list(filter(f1,p))
print(x)
```

### Creating a list of even squares using conditional expressions:

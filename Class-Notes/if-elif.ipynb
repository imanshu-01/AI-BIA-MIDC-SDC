```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
operator = str(input("Enter an operator ,(+, -, *, /,%,//): "))

if operator == "+":
    result = num1 + num2
    print("Addition of",num1, "and", num2, "is:",result)
    #print(f"Addition of {num1} and {num2} is: {result}")
elif operator == "-":
    result = num1 - num2
    print("Substraction of",num1 ,"and", num2, "is:",result)
elif operator == "*":
    result = num1 * num2
    print("Multiplication of",num1, "and", num2, "is:",result)
elif operator == "/":
    result = num1 / num2
    print("division of",num1,"and", num2, "is:",result)
elif operator == "%":
    result = num1 %num2
    print("Remainder of",num1,"and", num2, "is:",result)

elif operator == "//":
    result = num1 //num2
    print("Qoutient of",num1,"and", num2, "is:",result)

else:
    print("Invalid operator")
    

```

### 2.Student Grade checker

```python
percentage = float(input("Enter the percentage: "))

if percentage >= 90:
    grade = 'A'
elif percentage >= 80:
    grade = 'B'
elif percentage >=70:
    grade = 'C'
elif percentage >= 60:
    grade = 'D'
else:
    grade = 'E'

print("The student's grade is:", grade)
```

### 4. Marks checker

```python
marks=int(input("Enter the marks:"))

if 90<=marks<=100:
    print("Student got A+ grade")
elif 80<=marks<90:
    print("Student got A grade")
elif 70<=marks<80:
    print("student got B+ grade")
elif 60<=marks<70:
    print("student got  B grade")
elif 50<=marks<60:
    print("student got C+ grade")
elif 40<=marks<50:
    print("student got C grade")
else:
    print("student is fail")
```

## 5.Example code to classify temperature ranges

```python
temperature = float(input("Enter the temperature in Celsius: "))

if temperature <=0:
    classification = "Freezing"
elif 0 < temperature < 10:
    classification = "Cold"
elif 10 <= temperature < 20:
    classification = "Cool"
elif 20 <= temperature < 30:
    classification = "Warm"
else:
    classification = "Hot"

print("The temperature is classified as:", classification)
```

#### Build a program that reads a number from the user and identifies which range it falls into: 1-10, 11-20, 21-30, or above 30.

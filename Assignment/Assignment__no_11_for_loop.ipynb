13. Write a program to print a pyramid pattern of stars using nested for loops.

```python
# Program to print a pyramid pattern of stars

n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    # print spaces
    for j in range(n - i):
        print(" ", end="")
    # print stars
    for k in range(2 * i - 1):
        print("*", end="")
    print()
```

16. Implement a program to find all prime numbers in a given range using a for loop and the Sieve of Eratosthenes algorithm.

```python
# Program to find all prime numbers in a given range using Sieve of Eratosthenes

start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

# Step 1: Create a list of True values (prime candidates)
prime = [True] * (end + 1)
prime[0] = prime[1] = False   # 0 and 1 are not prime

# Step 2: Sieve algorithm using loops
for i in range(2, int(end**0.5) + 1):
    if prime[i]:  # If number is still True, it's prime
            for j in range(i * i, end + 1, i):
                        prime[j] = False   # Mark multiples as not prime

# Step 3: Print primes in the given range
print("Prime numbers between",start , "and", end, ":")
for num in range(start, end + 1):
  if prime[num]:
      print(num)
```

17. Build a program that generates the nth term of the arithmetic progression 2, 5, 8, 11, ... using a for loop.

```python
# Program to find the nth term of an arithmetic progression

n = int(input("Enter n: "))

a = 2   # first term
d = 3   # common difference

term = a

for i in range(1, n):
    term = term + d

    print("The", n, "th term of the AP is:", term)
```

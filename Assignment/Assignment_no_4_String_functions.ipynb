**String Functions**

1. Given the string text = "Hello, World!", use the len() function to find the length of the string.


```python
text = "Hello, World!"
length = len(text)
print("The length of the string is: {length}")
```

2. Create a string sentence "I am learning Python programming." and use the upper() function to convert it to uppercase.




```python
sentence = "I am learning Python programming."
uppercase_sentence = sentence.upper()
print("Original sentence: {sentence}")
print("Uppercase sentence: {uppercase_sentence}")
```

3. Using the count() function, find out how many times the letter "a" appears in the string "banana".


```python
word = "banana"
count_a = word.count("a")
print("The letter 'a' appears {count_a} times in the word '{word}'.")
```

4. Given the string message = "Welcome to Python!", use the strip() function to remove leading and trailing spaces.


```python
message = "  Welcome to Python!  "
stripped_message = message.strip()
print("Original message: '{message}'")
print("Stripped message: '{stripped_message}'")
```


5. Create a string word = "excellent" and use the replace() function to replace "cell" with "cent".


```python
word = "excellent"
new_word = word.replace("cell", "cent")
print("Original word: '{word}'")
print("New word: '{new_word}'")
```


6. Using the find() function, determine the index of the first occurrence of the substring "dog" in the string "The quick brown fox jumps over the lazy dog.".


```python
main_string = "The quick brown fox jumps over the lazy dog."
sub_string = "dog"
index = main_string.find(sub_string)
print("The string is: '{main_string}'")
print("The substring '{sub_string}' is found at index: {index}")
```

7. Given the string text "one, two, three, four, five", use the split() function to split it into a list of words.


```python
text = "one, two, three, four, five"
words = text.split(", ")
print("Original string: '{text}'")
print("List of words: {words}")
```


8. Create a string sentence = "Python is fun!" and use the startswith() function to check if it starts. with the word "Python"


```python
sentence = "Python is fun!"
starts_with_python = sentence.startswith("Python")
print("Does the sentence '{sentence}' start with 'Python'? {starts_with_python}")
```


9. Using the join() function, combine the elements of the list words=["Hello", "world", "Python"] into a sentence.


```python
words = ["Hello", "world", "Python"]
sentence = " ".join(words)
print("Original list of words: {words}")
print("Combined sentence: '{sentence}'")
```

10. Given the string phrase "Hello there, how are you?", use the capitalize() function to capitalize the first letter of the sentence.x.

```python
phrase = "Hello there, how are you?"
capitalized_phrase = phrase.capitalize()
print("Original phrase: '{phrase}'")
print("Capitalized phrase: '{capitalized_phrase}'")
```

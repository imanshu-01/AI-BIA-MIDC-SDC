# Set Exercises

1. Create a Python set and add the elements 1, 2, 3, 4, and 5 to it. Print the set.

```python
s = set()
s.update([1,2,3,4,5])  # add multiple elements
print("Set s:", s)
```

2. Write a Python program to remove an item from a set, and print the set before and after removal.

```python
s = {1,2,3,4,5}
print("Before removal:", s)
item_to_remove = 3
s.remove(item_to_remove)  # raises KeyError if item not present
print(f"After removing {item_to_remove}:", s)
```

3. Create two sets and use the union() method to combine them. Print the result.

```python
a = {1,2,3}
b = {3,4,5}
union_ab = a.union(b)  # or a | b
print("Set a:", a)
print("Set b:", b)
print("Union of a and b:", union_ab)
```

4. Write a Python program to find the intersection of two sets.

```python
a = {1,2,3,4}
b = {3,4,5,6}
intersection_ab = a.intersection(b)  # or a & b
print("Intersection of a and b:", intersection_ab)
```

5. Check if an element exists in a set using the `in` operator.

```python
s = {10,20,30,40}
x = 20
print(f"Is {x} in set s? ->", x in s)
y = 99
print(f"Is {y} in set s? ->", y in s)
```

6. Write a Python program to find the difference between two sets (elements in one set but not the other).

```python
a = {1,2,3,4,5}
b = {4,5,6,7}
diff_a_b = a.difference(b)  # elements in a but not in b
diff_b_a = b.difference(a)  # elements in b but not in a
print("a - b:", diff_a_b)
print("b - a:", diff_b_a)
```

7. Create a Python set and use the discard() method to remove an item. What happens if the item doesn't exist in the set?

```python
s = {1,2,3}
print("Original set:", s)
s.discard(2)  # removes 2
print("After discard(2):", s)
s.discard(99)  # does nothing if item not present (no error)
print("After discard(99) (item not present):", s)
# Explanation:
print("\nExplanation: discard() removes an element if present; it does NOT raise an error if the element is missing.")
```

8. Write a program to check if one set is a subset of another set.

```python
a = {1,2}
b = {1,2,3,4}
print("Is a subset of b?", a.issubset(b))  # True
print("Is b subset of a?", b.issubset(a))  # False
# alternative using <= operator
print("Using <= operator: a <= b ->", a <= b)
```

9. Create two sets of strings. Use the symmetric_difference() method to find elements that are in either set but not both.

```python
s1 = {"apple", "banana", "cherry"}
s2 = {"banana", "date", "fig"}
sym_diff = s1.symmetric_difference(s2)  # or s1 ^ s2
print("s1:", s1)
print("s2:", s2)
print("Symmetric difference (elements in either s1 or s2 but not both):", sym_diff)
```

10. Write a Python program that takes two sets and returns a set with the common elements (intersection) using set comprehension.

```python
a = {1,2,3,4}
b = {3,4,5,6}
intersection_comp = {x for x in a if x in b}
print("Intersection using set comprehension:", intersection_comp)
```

11. Create a Python program that removes all duplicate elements from a list using sets.

```python
lst = [1,9,8,2,2,3,7,4,4,4,5]
print("Original list:", lst)
unique_list = list(set(lst))  # order may change
print("List after removing duplicates (unordered):", unique_list)

# If order must be preserved:
seen = set()
unique_preserve = []
for item in lst:
    if item not in seen:
        unique_preserve.append(item)
        seen.add(item)
print("List after removing duplicates (order preserved):", unique_preserve)
```

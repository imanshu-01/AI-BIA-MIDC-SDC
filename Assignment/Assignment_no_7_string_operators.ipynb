1. Create a program that takes the base and height of a triangle as input and calculates its area (Â½ *
base * height).

```python
# Function to calculate the area of a triangle
def calculate_triangle_area(base, height):
    return 0.5 * base * height

# Get input from the user
base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))

# Calculate the area
area = calculate_triangle_area(base, height)

# Display the result
print(f"The area of the triangle with base {base} and height {height} is: {area}")
```

3. Write a program that simulates a simple calculator. The program should take two numbers(user
inputs)and an operator (+, -, *, /) as input and perform the corresponding operation, displaying
the result.

```python
# Get input from the user
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator (+, -, *, /): ")
num2 = float(input("Enter the second number: "))

# Perform the calculation based on the operator
if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero"
else:
    result = "Error: Invalid operator"

# Display the result
print(f"Result: {result}")
```

4. How do you concatenate two strings in Python?

```python
string1 = "Hello"
string2 = "World"

# Using the + operator
concatenated_string_plus = string1 + " " + string2
print(f"Using + operator: {concatenated_string_plus}")

# Using f-strings
concatenated_string_fstring = f"{string1} {string2}"
print(f"Using f-string: {concatenated_string_fstring}")
```

5. How can you convert an integer to a float, and vice versa?

```python
# Integer to Float
integer_num = 10
float_num = float(integer_num)
print(f"Integer {integer_num} converted to float: {float_num} (type: {type(float_num)})")

# Float to Integer
float_num2 = 10.75
integer_num2 = int(float_num2)
print(f"Float {float_num2} converted to integer: {integer_num2} (type: {type(integer_num2)})")

float_num3 = 9.2
integer_num3 = int(float_num3)
print(f"Float {float_num3} converted to integer: {integer_num3} (type: {type(integer_num3)})")
```

9. Create a program that takes a string and a character as input. Use the in and not in operators to
check if the character is present in the string and print the result.

```python
input_string = input("Enter a string: ")
input_char = input("Enter a character or substring to check: ")

# Check if the character is in the string using 'in'
is_present = input_char in input_string
print(f"Is '{input_char}' present in '{input_string}'? {is_present}")

# Check if the character is not in the string using 'not in'
is_not_present = input_char not in input_string
print(f"Is '{input_char}' not present in '{input_string}'? {is_not_present}")
```

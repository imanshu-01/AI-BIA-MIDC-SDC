4. How do you create an empty tuple?

Like an empty list, we can create an empty tuple in Python. Empty tuples can work as a placeholder and are more memory efficient compared to empty lists because tuples are immutable.

```python
# Creating an empty tuple using parentheses
t = ()
print(t)
print(type(t))
print(len(t))


# Creating an empty tuple using the tuple() constructor
t = tuple()
print(t)
print(type(t))
print(len(t))
```

8. How can you access a specific element in a tuple?

Elements within a tuple can be accessed using indexing, similar to how elements in a list are accessed. Python uses 0-based indexing, meaning the first element is at index 0, the second at index 1, and so on.

```python
# Positive Indexing: To access an element using positive indexing, specify the tuple name followed by the element's index enclosed in square brackets.

my_tuple = ("apple", "banana", "cherry")
first_element = my_tuple[0]  # Accesses "apple"
second_element = my_tuple[1] # Accesses "banana"
print(first_element)
print(second_element)
```

```python
# Negative indexing: It allows access to elements starting from the end of the tuple. The last element is at index -1, the second-to-last at index -2, and so forth.

my_tuple = ("apple", "banana", "cherry")
last_element = my_tuple[-1]  # Accesses "cherry"
second_last_element = my_tuple[-2] # Accesses "banana"
print(last_element)
print(second_last_element)
```

10. Write a Python code that takes a tuple of numbers as input and returns
the sum of those numbers.

```python
def sum_tuple_numbers(numbers_tuple):
    """
    Calculates the sum of numbers in a given tuple.

    Args:
        numbers_tuple (tuple): A tuple containing numbers.

    Returns:
        int or float: The sum of the numbers in the tuple.
    """
    return sum(numbers_tuple)

# Example usage:
my_numbers = (10, 20, 30, 40, 50)
total_sum = sum_tuple_numbers(my_numbers)
print(f"The tuple is: {my_numbers}")
print(f"The sum of the numbers in the tuple is: {total_sum}")

# Another example with floats:
another_tuple = (1.5, 2.5, 3.0)
float_sum = sum_tuple_numbers(another_tuple)
print(f"\nThe tuple is: {another_tuple}")
print(f"The sum of the numbers in the tuple is: {float_sum}")
```

11. Given a tuple (name, age, city), how would you unpack it into three separate variables?

You can unpack a tuple into separate variables by assigning the tuple to a comma-separated list of variable names on the left side of an assignment operator, ensuring the number of variables matches the number of elements in the tuple. For example, if you have a tuple person = ('Alice', 30, 'New York'), you would use the line name, age, city = person to unpack it.

**Unpacking syntax:** Place the variable names on the left of the equals sign and the tuple on the right.

**Matching count:** The number of variables must be equal to the number of items in the tuple. A mismatch will result in a ValueError.

**Example:**
person = ('Alice', 30, 'New York')
name, age, city = person
Now, name holds 'Alice', age holds 30, and city holds 'New York'.

```python
# Define the tuple
person_info = ("Alice", 30, "New York")

# Unpack the tuple into three separate variables
name, age, city = person_info

# Print the individual variables to verify
print(f"Name: {name}")
print(f"Age: {age}")
print(f"City: {city}")
```

20. Can you modify an element within a tuple after it has been created?
Why or why not?

No, you cannot directly modify an element within a tuple after it has been created.

**Reason:**
Tuples in Python are immutable. This means that once a tuple is defined and its elements are set, you cannot change, add, or remove any of those elements. Any attempt to modify a tuple directly will result in a TypeError.

```python
# Example of trying to modify a tuple element
my_immutable_tuple = (1, 2, 3, 'a', 'b')
print(f"Original tuple: {my_immutable_tuple}")

try:
    # Attempt to change the second element (at index 1) from 2 to 200
    my_immutable_tuple[1] = 200
    print(f"Modified tuple: {my_immutable_tuple}")
except TypeError as e:
    print(f"\nError: {e}")
    print("This error confirms that tuples are immutable and their elements cannot be changed.")

# You can, however, create a *new* tuple based on an existing one with modifications:
new_tuple_from_old = my_immutable_tuple[:1] + (200,) + my_immutable_tuple[2:]
print(f"\nNew tuple created from old with a change: {new_tuple_from_old}")
print(f"Original tuple remains unchanged: {my_immutable_tuple}")
```

21. If you have a list of names and want to ensure that it remains
unchanged, would you use a list or a tuple?

You should use a tuple to ensure a list of names remains unchanged because tuples are immutable. Lists are mutable, meaning their contents can be modified after creation, while tuples cannot be changed once they are created.

```python
# Using a tuple for names that should not change
immutable_names = ("Alice", "Bob", "Charlie")
print(f"Tuple of names: {immutable_names}")

try:
    # Attempting to change an element in the tuple will raise an error
    immutable_names[0] = "Alicia"
except TypeError as e:
    print(f"\nError when trying to modify tuple: {e}")
    print("Tuples prevent changes, ensuring immutability.")

# Using a list for names that can change
mutable_names = ["David", "Eve", "Frank"]
print(f"\nList of names (before change): {mutable_names}")

# Modifying an element in the list is allowed
mutable_names[0] = "Daniel"
mutable_names.append("Grace")
print(f"List of names (after change): {mutable_names}")
print("Lists allow changes, demonstrating mutability.")
```

25. Explain a scenario where using a list would be more appropriate than
using a tuple.

```python
# Initial inventory
inventory = ["Laptop", "Mouse", "Keyboard", "Monitor"]

# A customer buys a Mouse
inventory.remove("Mouse")

# New products arrive
inventory.append("Webcam")
inventory.append("Headphones")

# A product is temporarily out of stock and needs to be replaced with an alternative
inventory[inventory.index("Keyboard")] = "Ergonomic Keyboard"

print(inventory)
```

In this scenario, the inventory needs to be dynamically modified:

* Adding new items: When new products are stocked.
* Removing items: When products are sold.
* Modifying existing items: When a product is replaced or updated.

1. Identify if a triangle is equilateral, isosceles, or scalene based on side lengths.

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

# Check if the sides can form a valid triangle
if a + b > c and a + c > b and b + c > a:

    if a == b == c:
        print("The triangle is Equilateral.")
    elif a == b or b == c or a == c:
        print("The triangle is Isosceles.")
    else:
        print("The triangle is Scalene.")

else:
    print("The given sides do not form a valid triangle.")
```

2. Find the largest among three numbers using nested ifs.

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

if a > b:
    if a > c:
        print("Largest number is:", a)
    else:
        print("Largest number is:", c)
else:
    if b > c:
        print("Largest number is:", b)
    else:
        print("Largest number is:", c)
```

3. Check if a person is eligible to vote based on age and nationality.

```python
age = int(input("Enter age: "))
nationality = input("Enter nationality: ")

if age >= 18:
    if nationality.lower() == "indian":
        print("The person is eligible to vote.")
    else:
        print("Not eligible: Nationality must be Indian.")
else:
    print("Not eligible: Age must be 18 or above.")
```

4. Check if a user can withdraw a certain amount based on balance and limits.

```python
balance = float(input("Enter your account balance: "))
withdraw = float(input("Enter withdrawal amount: "))
limit = float(input("Enter withdrawal limit: "))

if withdraw <= limit:
    if withdraw <= balance:
        print("Withdrawal Successful!")
    else:
        print("Withdrawal Failed: Insufficient Balance.")
else:
    print("Withdrawal Failed: Amount exceeds withdrawal limit.")
```

5. Check if a student passed all subjects, and then determine grade based on average.

```python
# Taking marks of 5 subjects
marks = []

for i in range(1, 5 + 1):
    m = float(input(f"Enter marks of subject {i}: "))
    marks.append(m)

# Check if the student passed in every subject
if all(m >= 35 for m in marks):
    avg = sum(marks) / len(marks)

    # Determine grade based on average
    if avg >= 90:
        grade = "A"
    elif avg >= 75:
        grade = "B"
    elif avg >= 60:
        grade = "C"
    else:
        grade = "D"

    print("\nResult: PASSED")
    print("Average Marks:", avg)
    print("Grade:", grade)

else:
    print("\nResult: FAILED (One or more subjects are below 35)")
```

6. Simulate a login process by checking username and password using nested ifs.

```python
# Saved / correct login details
correct_username = "admin"
correct_password = "1234"

# User input
username = input("Enter username: ")
password = input("Enter password: ")

# Nested if to check login
if username == correct_username:
    if password == correct_password:
        print("Login Successful! Welcome.")
    else:
        print("Login Failed: Incorrect Password.")
else:
    print("Login Failed: Incorrect Username.")
```

7. Check if a person is eligible to donate blood based on age and weight.

```python
age = int(input("Enter age: "))
weight = float(input("Enter weight (in kg): "))

if age >= 18:
    if weight >= 50:
        print("Eligible to donate blood.")
    else:
        print("Not eligible: Weight must be at least 50 kg.")
else:
    print("Not eligible: Age must be 18 or above.")
```

```python
8. Apply discounts based on age (children, seniors) and if they have a discount card.
```

```python
age = int(input("Enter age: "))
has_card = input("Do you have a discount card? (yes/no): ").lower()

discount = 0

# Age-based discounts
if age < 12:
    discount = 50      # 50% discount for children
elif age >= 60:
    discount = 30      # 30% discount for seniors
else:
    discount = 0       # No age-based discount

# Extra discount for card holders
if has_card == "yes":
    discount += 10     # Additional 10%

print("Total Discount =", discount, "%")
```

9. Calculate bonus based on years of service and job role.

```python
years = int(input("Enter years of service: "))
role = input("Enter job role (manager/staff): ").lower()

if years > 5:
    if role == "manager":
        print("Bonus = 20%")
    else:
        print("Bonus = 10%")
else:
    print("No bonus (service must be more than 5 years).")
```

10. Check if a person is eligible for a loan based on income and credit score.

```python
income = float(input("Enter monthly income: "))
credit_score = int(input("Enter credit score: "))

if income >= 30000:
    if credit_score >= 700:
        print("Loan Approved!")
    else:
        print("Loan Denied: Credit score is too low.")
else:
    print("Loan Denied: Income must be at least 30000.")
```

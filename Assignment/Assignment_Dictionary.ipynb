# Dictionary & Frozenset Exercises

```python
# 1. Create a dictionary called person with keys name, age, and city.
# Initialize the dictionary with your personal details. Print name and city.
person = {"name": "Himanshu", "age": 24, "city": "Nagpur"}
print("Name:", person["name"])
print("City:", person["city"])
```

```python
# 2. Given fruits dict, increase banana by 3 and remove orange. Print updated dict.
fruits = {'apple': 2, 'banana': 5, 'orange': 8}
fruits['banana'] += 3
fruits.pop('orange', None)
print("Updated fruits:", fruits)
```

```python
# 3. Check if key 'name' exists in dictionary and print value if it does.
data = {'name': 'John', 'age': 25, 'city': 'New York'}
if 'name' in data:
    print("Key 'name' exists. Value:", data['name'])
else:
    print("Key 'name' not found.")
```

```python
# 4a. Use pop() to remove Charlie and return his grade.
# 4b. Use keys(), values(), items() to print keys, values, and key-value pairs.
student_grades = {'Alice': 95, 'Bob': 85, 'Charlie': 90}
charlie_grade = student_grades.pop('Charlie', None)
print("Charlie's grade (popped):", charlie_grade)
print("Keys:", list(student_grades.keys()))
print("Values:", list(student_grades.values()))
print("Items:", list(student_grades.items()))
```

```python
# 5. Create a dictionary where each key maps to a list of subjects.
# Retrieve all subjects for a given student (example: 'Alice').
student_subjects = {
    'Alice': ['Math', 'Physics'],
    'Bob': ['History', 'Geography'],
    'Charlie': ['Biology', 'Chemistry']
}
student = 'Alice'
subjects = student_subjects.get(student, [])
print(f"Subjects for {student}:", subjects)
```

```python
# 6. Create a frozenset containing 10,20,30,40,50 and check if 30 present.
fs = frozenset([10, 20, 30, 40, 50])
print("Frozenset:", fs)
print("Is 30 present?", 30 in fs)
```

```python
# 7. Try to add an element to a frozenset and explain the result.
fs = frozenset([1, 2, 3])
try:
    # frozenset has no add() method; this will raise AttributeError if attempted
    fs.add(4)  # intentionally causes an error
except Exception as e:
    print("Error when trying to add to frozenset:", type(e).__name__, "-", e)
print("\nExplanation: frozenset is immutable; you cannot modify it after creation. To get a new frozenset with an extra element, use union:")
fs_new = fs.union([4])
print("New frozenset via union:", fs_new)
```

```python
# 8. Create fs1 and fs2 and compute union, intersection, and difference.
fs1 = frozenset([1, 2, 3, 4])
fs2 = frozenset([3, 4, 5, 6])
print("fs1:", fs1)
print("fs2:", fs2)
print("Union:", fs1.union(fs2))
print("Intersection:", fs1.intersection(fs2))
print("Difference (fs1 - fs2):", fs1.difference(fs2))
```
